```python
from datasets import load_dataset
import pandas as pd
```

```python
from pathlib import Path

# Загрузка датасета:
dataset = load_dataset("d0rj/geo-reviews-dataset-2023")
```

```python
# Преобразование данных в DataFrame:

data_df = pd.DataFrame(dataset['train'])
```

```python
data_df.info()
```

```python
data_df.head()
```

```python
# Удаление пропущенных значений

work_data = data_df.dropna(subset=['text', 'name_ru', 'rating'])
```

```python
# Удаление дубликатов

work_data = work_data.drop_duplicates(subset=['text']).reset_index(drop=True)
```

```python
# Замена переносов строк

work_data['text'] = work_data['text'].str.replace('\\n', ' ')
```

```python
# Ограничение объема данных

work_data = work_data[:15000]
```

```python
work_data.info()
```

```python
work_data.head()
```

```python
# Сохраняем очищенные данные в новый файл
work_data.to_csv('work_data.csv', index=False)

print("\nДанные успешно очищены и сохранены в 'work_data.csv'.")
```
